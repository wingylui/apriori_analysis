```python
import pandas as pd
import numpy as np
```

```python
with open("./week1_assign.txt", "r") as file: 
    data = file.readlines()
    dum = []
    file_set = set()
    for row in data:
        dum.append(row.strip().split(";"))
        file_set.update(row.strip().split(";"))

df = pd.DataFrame(dum)

df.shape
```

```python
ls_dum = []
for index, row in df.iterrows():
    dum = set(row.values)
    ls_dum.append(dum)
len(ls_dum)
```

```python
# create blank list in the dict
unique_cat = {k: [] for k in file_set} 
for index, row in df.iterrows():
    dum = list(row.values)
    # remove all the items find in the dum list
    set_ = file_set.difference(dum)
    # for loop the dum list and append the key into each list and others not find in dum list will loop through set_ and insert null value
    for key in dum:
        if key != None:
            unique_cat[key].append(True)
        else: 
            break
    for key in set_:
        unique_cat[key].append(False)


unique_cat["Restaurants"]
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Extracting the most frequest itemsets via Mlxtend.
# The length column has been added to increase ease of filtering.
frequent_itemsets = apriori(final_df, min_support=0.01, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets["abs support"] = round(frequent_itemsets["support"] * df.shape[0]).astype(int)
# printing the frequent itemset
frequent_itemsets
```

```python
## convert the forzenset into list
item_list  = []
abs_support_list = []
for index, row in frequent_itemsets.iterrows():
    item_list.append(list(row["itemsets"]))
    abs_support_list.append(row["abs support"])
```

```python
result_string = str()
for n in range(len(item_list)):
    if len(item_list[n]) == 1:
        item_str = str()
        result_string = str(result_string) + f"{int(abs_support_list[n])}:{item_list[n][0]}\n"
result_string = result_string.strip()
result_string
```

```python
with open('./week1_assign_results.txt', 'w') as f:
    f.writelines(result_string)
```
